### Part B – Determine the date span of the data

Write Python statements to query the daily_measures table in the dj.db database, to determine the
earliest and latest date of the data in this table

```python
# place SQL results in a DataFrame - method 2: pandas read_sql_query()
# use named parameter index_col
# Use SQL to join 2 tables; Python statements are similar to one-table queries
import sqlite3 as sql
connection = sql.Connection("dj.db")
cursor = connection.cursor()
query = """
 SELECT min(date) as first, max(date) as last FROM daily_measures
"""
cursor.execute(query)
rows = cursor.fetchall()
print(rows)
```

### Answer: The earliest date in this data is January 3rd, 2017, and the last data is December 29th, 2017. 

## Part C – Query price history with for and Cursor

Write Python statements to query the daily_measures table in the dj.db database, and based on the
returned data, determine the date and closing price of the highest price for Apple stock during the data
available in the database.

```python
# fetchone() used here with for
import sqlite3 as sql
connection = sql.Connection("dj.db")
cursor = connection.cursor()
cursor.execute(" SELECT date, closing FROM daily_measures WHERE ticker_id=3 ORDER BY date")
for row in cursor: # Cursor provides a shorthand way to iterate over the rows
   print(row)
```

```python
#Find the date of Apple's maximum closing price
import sqlite3 as sql
connection = sql.Connection("dj.db")
cursor = connection.cursor()
cursor.execute(" SELECT date, max(closing) FROM daily_measures WHERE ticker_id=3 ORDER BY date")
for row in cursor: # Cursor provides a shorthand way to iterate over the rows
   print(row)
```

### Answer: Apple's max closing price was 176.42 dollars, on December 18, 2017.
